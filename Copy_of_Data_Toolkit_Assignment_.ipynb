Q(14)Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
#Ans:-
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {'Date': pd.date_range(start='2023-01-01', end='2023-01-10'),
        'Temperature': [25, 27, 26, 28, 24, 26, 27, 25, 26, 27],
        'Humidity': [60, 65, 62, 68, 58, 63, 66, 61, 64, 67]}

df = pd.DataFrame(data)

# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], color='red')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params('y', colors='red')

# Create the right y-axis
ax2 = ax1.twinx()

# Plot humidity on the right y-axis
ax2.plot(df['Date'], df['Humidity'], color='blue')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params('y', colors='blue')

# Set the title
plt.title('Temperature and Humidity Over Time')

plt.show()



```

Q(15) Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib:

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1

# Step 2: Plot the histogram
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Step 3: Overlay the PDF of the normal distribution
mean = np.mean(data)
std_dev = np.std(data)
x = np.linspace(min(data), max(data), 100)
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'k', linewidth=2)

# Step 4: Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Step 5: Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Show the plot
plt.grid()
plt.show()
```

Q(18)Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'.

```python
#Ans:-
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource, HoverTool

# Step 1: Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame for better handling
data = pd.DataFrame({'Category': categories, 'Value': values})

# Step 2: Create a ColumnDataSource
source = ColumnDataSource(data)

# Step 3: Set up the output file (this will create an HTML file)
output_file("random_categorical_bar_chart.html")

# Step 4: Create a figure
p = figure(x_range=data['Category'], title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values', toolbar_location=None)

# Step 5: Add a bar renderer with color based on values
p.vbar(x='Category', top='Value', source=source,
        line_color='white', fill_color='blue')

# Step 6: Add hover tooltips to display exact values
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

# Step 7: Customize the grid and show the plot
p.y_range.start = 0
p.xgrid.grid_line_color = None
p.ygrid.grid_line_alpha = 0.5

# Step 8: Show the plot
show(p)



```

Q(20)Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set
the title as 'Interactive Pie Chart'.

```python
#Ans:-
import numpy as np
import plotly.express as px

# Step 1: Generate random data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
values = np.random.randint(1, 100, size=len(labels))

# Step 2: Create a pie chart using Plotly Express
fig = px.pie(values=values, names=labels, title='Interactive Pie Chart',
             hover_data=[values], labels={label: f"{label}: {value}" for label, value in zip(labels, values)})

# Step 3: Update the traces to show percentages
fig.update_traces(textinfo='percent+label')

# Step 4: Show the plot
fig.show()






```
